## Libraries -
1. beautifulsoup - https://pypi.org/project/beautifulsoup4/
2. lxml - https://pypi.org/project/lxml/

```python
from bs4 import BeautifulSoup
import requests
import csv
import lxml

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

category_dic = {"Laptop":"https://www.amazon.in/s?k=laptop&page=","Smartphone":"https://www.amazon.in/s?k=Smartphone&page="}




Product_Name_List = []
Price_List = []
Offer_Price_List = []
Original_Price_List = []
Product_Link = []



for page in range(0,5):
    source = requests.get(category_dic["Laptop"] + str(page+1), headers = headers).text
    soup = BeautifulSoup(source, "lxml")

    #print(soup.prettify())

    

    # finding product name
    for name in soup.find_all('a', class_='a-link-normal a-text-normal'):
        str_name = name.text
        Product_Name_List.append(str_name.strip())



    # finding price
    for price in soup.find_all('span', class_="a-offscreen"):
        str_price = price.text
        Price_List.append(str_price.strip())

    for counter1 in range(0, len(Price_List), 2):
        Offer_Price_List.append(Price_List[counter1])

    for counter2 in range(1, len(Price_List), 2):
        Original_Price_List.append(Price_List[counter2])



    # finding product link
    for link in soup.find_all('a', class_='a-link-normal a-text-normal'):
        str_link = link['href']
        Product_Link.append("https:amazon.in//{}".format(str_link.strip()))

```

```python
# printing Name, offer price, original price and product link

for x in range(0,len(Product_Name_List)):
    print("Sr No-",x+1)
    print("Product Name:\n",(Product_Name_List[x]))
    print("\nOffer Price:\n",(Offer_Price_List[x]))
    print("\nOriginal Price:\n",(Original_Price_List[x]))
    print("\nProduct Link:\n",(Product_Link[x]))
    print("\n\n\n")
```
